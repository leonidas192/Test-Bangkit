```python
import matplotlib.pyplot as plt
fiq,ax=plt.subplots()
ax.scatter(iris['sepal_length'],iris['sepal_width'])
                #x                        #y           
ax.set_title('iris dataset')
ax.set_xlabel('sepal_length')#untuk sumbu x
ax.set_ylabel('sepal_width')#untuk sumbu y`
```

knowledge :

Scatter plot atau diagram pencar atau juga disebut diagram sebar adalah gambaran yang menunjukkan
kemungkinan hubungan (korelasi) antara dua variabel dan menunjukkan keeratan hubungan antara dua
variabel tersebut.

clustering adalah sebuah proses untuk mengelompokan data ke dalam beberapa cluster atau kelompok sehingga data dalam satu cluster memiliki tingkat kemiripan yang maksimum dan data antar cluster memiliki kemiripan yang minimum.

keterangan ouput:

kemungkinan dataset tersebut makin panjang sepalnya maka makin lebar sepalnya , tetapi masih terlalu abstrak untuk melihat pengelompokannya secara detail untuk itu membutuh cluster untuk membedakannnya dengan memberi pewarnaan di tiap group agar data dapat terlihat lebih visualisasi dan 





```python
import matplotlib.pyplot as plt

colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}#dictionary color


fiq,ax=plt.subplots()#dictionary

for i in range(len(iris['sepal_length'])):#fungsi length singkatan len merupakan panjang variabel membaca index terakhir 
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],color=colors[iris['class'][i]])
        
ax.set_title('iris dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
import matplotlib.pyplot as plt

colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}


fiq,ax=plt.subplots()

for i in range(len(iris['petal_length'])): 
    ax.scatter(iris['petal_length'][i], iris['petal_width'][i],color=colors[iris['class'][i]])
        
ax.set_title('iris dataset')
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')
```

```python
import matplotlib.pyplot as plt

from numpy import sin
# consistent interval for x-axis
x = [x*0.1 for x in range(100)]
# function of x for y-axis
y = sin(x)
# createline plot
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
wine_reviews = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)


fig, ax = plt.subplots()
ax.hist(wine_reviews['points'])

ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

keterangan output:
memiliki median  yang sama sehingga terjadi kestabilan 

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
data = wine_reviews['points'].value_counts()
points = data.index
frequency = data.values
ax.bar(points, frequency)

ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

keterangan output:
memiliki kestasbilan data dengan adanya median
